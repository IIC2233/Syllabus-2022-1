## Manejo de excepciones: `try` y `except`

Las sentencias `try` y `except` nos permiten lidiar con los errores que pueden ocurrir en nuestro programa.

- `try`: permite definir un bloque de c贸digo en el que vamos a **intentar** ejecutar ciertas instrucciones. Si se **levanta** una excepci贸n dentro de este bloque entonces la excepci贸n es **capturada**.
- `except`: permite definir las instrucciones para manejar la excepci贸n capturada.

Sintaxis:
```python
try:
    # Lineas de codigo
except:
    # Se ejecuta si se levanta una excepcion dentro de try
```

### 驴C贸mo funciona?

1. Se ejecuta el c贸digo dentro de `try`.
2. Si no se levanta una excepci贸n, se termina de ejecutar este bloque y se sigue con las instrucciones definidas despu茅s del bloque `except`.
3. Si se levanta una excepci贸n, se dejan de ejecutar las instrucciones definidas en `try`y se comienzan a ejecutar las instrucciones del bloque `except`.

`````python
def dividir(numerador, denominador):
    check = isinstance(numerador, int) and isinstance(denominador, int)
    if not check:
        raise TypeError('Ambos argumentos deben ser tipo int')
    elif denominador == 0:
        raise ZeroDivisionError('No se puede dividir por 0')
    else:
        return numerador / denominador

try:
    resultado = dividir(1, "3")
except TypeError as error:
    print(f"Se levant贸 un TypeError: {error}")
except ZeroDivisionError as error:
    print(f"Se levant贸 un ZeroDivisionError: {error}")
`````

## Flujos complementarios: `else` y `finally`

Los bloques `try` y `except` pueden ser complementados **opcionalmente** con estas sentencias:
- `else`: Las instrucciones dentro de este bloque se ejecutar谩n siempre y cuando **NO** se haya levantado ninguna excepci贸n.
- `finally`: En este bloque van instrucciones que se realizar谩n **SIEMPRE**, independientemente de si ocurri贸 una excepci贸n o no.

Sintaxis:
```python
try:
    # Lineas de codigo
except:
    # Se ejecuta si se levanta una excepcion dentro de try
else:
    # Se ejecuta si no se levanta una excepcion dentro de try
finally:
    # Se ejecuta SIEMPRE
````

`````python
def dividir(numerador, denominador):
    check = isinstance(numerador, int) and isinstance(denominador, int)
    if not check:
        raise TypeError('Ambos argumentos deben ser tipo int')
    elif denominador == 0:
        raise ZeroDivisionError('No se puede dividir por 0')
    else:
        return numerador / denominador

try:
    resultado = dividir(1, 3)
except TypeError as error:
    print(f"Se levant贸 un TypeError: {error}")
except ZeroDivisionError as error:
    print(f"Se levant贸 un ZeroDivisionError: {error}")
else:
    print(f"El resultado de la divisi贸n es: {resultado}")
finally:
    print("Adiooooooooos")
`````
